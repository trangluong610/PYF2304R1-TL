```python
import tkinter as tk
from tkinter import messagebox
from tkinter.ttk import Treeview


window = tk.Tk()
window.title("Quản lý Nhà sách")


books = []


def add_book():
    book_name = entry_book_name.get()
    book_genre = entry_book_genre.get()
    book_price = entry_book_price.get()

    if book_name and book_genre and book_price:
        book = (book_name, book_genre, book_price)
        books.append(book)
        messagebox.showinfo("Thông báo", "Đã thêm sách thành công!")
    else:
        messagebox.showerror("Lỗi", "Vui lòng điền đầy đủ thông tin sách.")

   
    entry_book_name.delete(0, tk.END)
    entry_book_genre.delete(0, tk.END)
    entry_book_price.delete(0, tk.END)


def search_book():
    book_name = entry_search.get()
    found_books = []

    if book_name:
        for book in books:
            if book_name.lower() in book[0].lower():
                found_books.append(book)

        if found_books:
            display_books(found_books)
        else:
            messagebox.showerror("Thông báo", "Không tìm thấy sách.")
    else:
        messagebox.showerror("Lỗi", "Vui lòng nhập tên sách để tìm kiếm.")

    entry_search.delete(0, tk.END)

def delete_book():
    selected_item = treeview.focus()

    if selected_item:
        confirmation = messagebox.askyesno("Xác nhận", "Bạn có chắc chắn muốn xóa sách này?")
        if confirmation:
            treeview.delete(selected_item)
            for book in books:
                if book[0] == treeview.item(selected_item)["values"][0]:
                    books.remove(book)
                    break
            messagebox.showinfo("Thông báo", "Đã xóa sách thành công!")
    else:
        messagebox.showerror("Lỗi", "Vui lòng chọn sách để xóa.")

def update_price():
    selected_item = treeview.focus()

    if selected_item:
        new_price = entry_update_price.get()
        if new_price:
            treeview.set(selected_item, column="price", value=new_price)
            for book in books:
                if book[0] == treeview.item(selected_item)["values"][0]:
                    book = (book[0], book[1], new_price)
                    break
            messagebox.showinfo("Thông báo", "Đã cập nhật giá tiền sách!")
        else:
            messagebox.showerror("Lỗi", "Vui lòng nhập giá tiền mới.")
    else:
        messagebox.showerror("Lỗi", "Vui lòng chọn sách để cập nhật giá tiền.")

    entry_update_price.delete(0, tk.END)


def display_books(book_list):
    treeview.delete(*treeview.get_children())
    for book in book_list:
        treeview.insert("", tk.END, values=book)


label_book_name = tk.Label(window, text="Tên sách:")
label_book_name.grid(row=0, column=0, sticky="w")
entry_book_name = tk.Entry(window)
entry_book_name.grid(row=0, column=1)

label_book_genre = tk.Label(window, text="Thể loại:")
label_book_genre.grid(row=1, column=0, sticky="w")
entry_book_genre = tk.Entry(window)
entry_book_genre.grid(row=1, column=1)

label_book_price = tk.Label(window, text="Giá tiền:")
label_book_price.grid(row=2, column=0, sticky="w")
entry_book_price = tk.Entry(window)
entry_book_price.grid(row=2, column=1)

button_add = tk.Button(window, text="Thêm sách", command=add_book)
button_add.grid(row=3, column=0)

label_search = tk.Label(window, text="Tìm kiếm:")
label_search.grid(row=4, column=0, sticky="w")
entry_search = tk.Entry(window)
entry_search.grid(row=4, column=1)
button_search = tk.Button(window, text="Tìm kiếm", command=search_book)
button_search.grid(row=4, column=2)

treeview = Treeview(window, columns=("name", "genre", "price"), show="headings")
treeview.heading("name", text="Tên sách")
treeview.heading("genre", text="Thể loại")
treeview.heading("price", text="Giá tiền")
treeview.grid(row=5, column=0, columnspan=3)

button_delete = tk.Button(window, text="Xóa sách", command=delete_book)
button_delete.grid(row=6, column=0)

label_update_price = tk.Label(window, text="Giá tiền mới:")
label_update_price.grid(row=7, column=0, sticky="w")
entry_update_price = tk.Entry(window)
entry_update_price.grid(row=7, column=1)
button_update_price = tk.Button(window, text="Cập nhật giá tiền", command=update_price)
button_update_price.grid(row=7, column=2)


window.mainloop()
```
