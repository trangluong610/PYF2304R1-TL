```python
class Member:
    def __init__ (self, name, age, gender, BMI, card_type, number):
        self.name = name
        self.age = age
        self.gender = gender
        self.BMI = BMI
        self.card_type = card_type
        self.number = number
    
    def __str__(self):
        return f"Name: {self.name} \nAge: {self.age} \nGender: {self.gender} \nBMI: {self.BMI} \nCard type: {self.card_type} \nNumber: {self.number}"
```

```python
from tkinter import *
from tkinter import messagebox
from tkinter import ttk
from openpyxl import Workbook, load_workbook
import csv


window = Tk()
window.title("Member Gym Club Management")
window.geometry("500x400")


member_list = []

card_types = [
    'Gold',
    'Silver',
    'Classic'
    ]  


# Lấy dữ liệu từ tệp Excel và hiển thị trên Treeview
def load_data():
    for row in sheet.iter_rows(values_only=True):
        tree.insert("", "end", values=row)
        
# Load and save CSV
def load_csv():
    with open("new.csv") as myfile:
        csvread = csv.reader(myfile, delimiter=',')
        
        for row in csvread:
            print('load row:', row)
            tree.insert("", 'end', values=row)

def save_csv():
    with open("new.csv", "w", newline='') as myfile:
        csvwriter = csv.writer(myfile, delimiter=',')
        
        for row_id in tree.get_children():
            row = tree.item(row_id)['values']
            print('save row:', row)
            csvwriter.writerow(row)


# Entry(window)

name = Entry(window)
age = Entry(window)
gender = Entry(window)
search_member = Entry(window)
BMI = Entry(window)
number = Entry(window)
search_number = Entry(window)

def show_search_member():
    label.config( text = clicked.get() )
    
clicked = StringVar()

card_type_list = OptionMenu( window, clicked, *card_types )

button = Button(window, text = "click me" , command = show_search_member)

label = Label(window, text = " ")

def add_member():
    member_name = name.get()
    member_age = age.get()
    member_gender = gender.get()
    member_BMI = BMI.get()
    member_card_type = clicked.get()
    member_number = number.get()
  
    
    if member_name == "" or member_age == "" or member_gender == "" or member_BMI == "" or member_card_type == "" or member_number == "":
        messagebox.showerror("Add a member", "Please input all information")
    else:
        new_member = Member(member_name, member_age, member_gender, member_BMI, member_card_type, member_number)
        member_list.append(new_member)
        tree.insert("", "end", values=(member_name, member_age, member_gender, member_BMI, member_card_type, member_number))
        
        print([member.name and member.number for member in member_list])
        messagebox.showinfo("Add a member", f"Added {member_name} successfully!")

def search_member():
    search_number_data = search_number.get()
    check = False
    
    for i in range(len(member_list)):
        if member_list[i].number == search_number_data:
            messagebox.showinfo("Search a member", member_list[i])
            check = True
            break
    if check == False:
        messagebox.showerror("Search a member", "Cannnot find this member! Please try again")
        
def delete_member():
    selected_item = tree.selection()[0]
    tree.delete(selected_item)

tree = ttk.Treeview(window)
tree.pack(pady=6)
tree["column"] = ("name", "age", "gender", "BMI", "card_type", "number")

tree.column("name", width = 70)
tree.column("age", width = 70)
tree.column("gender", width = 70)
tree.column("BMI", width = 70)
tree.column("card_type", width = 70)
tree.column("number", width = 140)

tree.heading("name", text = "Name")
tree.heading("age", text = "Age")
tree.heading("gender", text = "Gender")
tree.heading("BMI", text = "BMI")
tree.heading("card_type", text = "Card type")
tree.heading("number", text = "Number")
        
        

lbl_add_member = Label(window, text = "ADD A MEMBER", font=("Arial", 11, "bold"))
lbl_name = Label(window, text = "Name: ", font=("Arial", 10))
lbl_age = Label(window, text = "Age: ", font=("Arial", 10))
lbl_gender = Label(window, text = "Gender: ")
lbl_BMI = Label(window, text = "BMI: ", font=("Arial", 10))
lbl_card_type = Label(window, text = "Card type: ", font=("Arial", 10))
lbl_registration_date = Label(window, text = "Registration date: ")
lbl_search_member = Label(window, text = "SEARCH A MEMBER", font=("Arial", 11, "bold"))
lbl_number = Label(window, text = "Number: ", font=("Arial", 10))
lbl_search_number = Label(window, text = "Input \nnumber: ", font=("Arial", 10))



btn_add = Button(window, text = "Add", command = add_member, bg = "#4CAF50", fg = "white", font = ("Arial", 10, "bold"))
btn_delete = Button(window, text = "Delete", command = delete_member, bg = "#F44336", fg = "white", font = ("Arial", 10, "bold"))
btn_search = Button(window, text = "Search", command = search_member, bg = "#4CAF50", fg = "white", font = ("Arial", 10, "bold"))
btn_load_csv = Button(window, text = "Load", command = load_csv, bg = "#2196F3", fg = "white", font = ("Arial", 10, "bold"))
btn_save_csv = Button(window, text = "Save", command = save_csv, bg = "#2196F3", fg = "white", font = ("Arial", 10, "bold"))

lbl_add_member.place(x=170, y=15)
lbl_name.place(x=80, y=50)
name.place(x=150, y=50)
lbl_age.place(x=80, y=75)
age.place(x=150, y=75)
lbl_gender.place(x=80, y=100)
gender.place(x=150, y=100)
lbl_BMI.place(x=80, y=125)
BMI.place(x=150, y=125)
lbl_card_type.place(x=80, y=150)
card_type_list.place(x=150, y=150)
lbl_number.place(x=80, y=190)
number.place(x=150, y=190)
btn_add.place(x=300, y=100)
btn_load_csv.place(x=1130, y=100)
btn_save_csv.place(x=1130, y=140)
btn_delete.place(x=1125, y=180)

lbl_search_member.place(x=160, y=235)
lbl_search_number.place(x=80,y=260)
search_number.place(x=150, y=270)
btn_search.place(x=290, y=265)


window.mainloop()
```
